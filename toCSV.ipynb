```python
#Useful function
def createFileList(myDir, format='.png'):
    fileList = []
    print(myDir)
    for root, dirs, files in os.walk(myDir, topdown=False):
        for name in files:
            if name.endswith(format):
                fullName = os.path.join(root, name)
                print(fullName)
                fileList.append(fullName)
    return fileList
```

```python
def resizeFile(imageFile):
    #print (imageFile)
    statinfo = os.stat(imageFile)
    if (statinfo.st_size < 2000):
        pic = Image.open(imageFile).convert('LA')

        width = 32
        height = 32
        img = pic.resize((width, height), Image.ANTIALIAS)
        img.save(imageFile)
    else:
        os.remove(imageFile)
```

```python
columnNames = list()

for i in range(1024):
    pixel = 'pixel'
    pixel += str(i)
    columnNames.append(pixel)

train_data = pd.DataFrame(columns = columnNames)
label_data = pd.DataFrame(columns = ['label'])
```

```python
myFileList = createFileList(myPath)
i=0
for file in myFileList:
    img = Image.open(file).convert('L')
    rawData = img.load()
    data = []
    for y in range(32):
        for x in range(32):
            data.append(rawData[x,y])
    k = 0
    train_data.loc[i] = [data[k] for k in range(1024)]
    i=i+1

print (i)

fileNameTrain = "char_" + str(label) + ".csv"
train_data.to_csv(fileNameTrain, index = False)

for j in range(i):
    label_data.loc[j] = label
fileNameLabel = "charLabels_" + str(label) + ".csv"
label_data.to_csv(fileNameLabel, index = False)
```
